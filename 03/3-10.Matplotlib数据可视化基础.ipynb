# matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np
X = np.linspace(0, 10, 100)
```

```python
X
```

```python
y = np.sin(X)
```

```python
y
```

```python
plt.plot(X, y)
```

```python
plt.show()
```

```python
cosy = np.cos(X)
```

```python
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(X, siny)
plt.plot(X, cosy)
plt.show()
```

```python
plt.plot(X, siny)
plt.plot(X, cosy, color='r')
plt.show()
```

```python
plt.plot(X, siny)
plt.plot(X, cosy, color='r', linestyle='--')
plt.show()
```

```python
plt.plot(X, siny)
plt.plot(X, cosy, color='r', linestyle=':')
plt.show()
```

```python
plt.plot(X, siny)
plt.plot(X, cosy, color='r', linestyle=':')
plt.xlim(-5, 15,)
plt.ylim(0,1.5)
plt.show()
```

```python
plt.plot(X, siny)
plt.plot(X, cosy, color='r', linestyle=':')
plt.axis([-1,15,-2,2])
plt.show()
```

```python
plt.plot(X, siny)
plt.plot(X, cosy, color='r', linestyle=':')
plt.xlabel("x axis")
plt.ylabel("y value")
plt.show()
```

```python
plt.plot(X, siny,label="sin(x)")
plt.plot(X, cosy, color='r', linestyle=':',label='cos(x)')
plt.xlabel("x axis")
plt.ylabel("y value")
plt.title('welcome to the ML World!')
plt.legend()
plt.show()
```

### Scatter Plot

```python
plt.scatter(X, siny)
plt.scatter(X, cosy, color='red')
plt.show()
```
